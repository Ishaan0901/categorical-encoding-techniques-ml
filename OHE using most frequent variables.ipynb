OneHotEncoding Usning most frequent variables:


```python
#Importing Libraries:
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#importing dataframe:
df=pd.read_csv('cars.csv')
df.head(2)
```

```python
counts=df['brand'].value_counts()
```

I will make a Threshold. If number of cars of the brand is less than the threshold , then i will club the values into a new column called 'Others'.

```python
threshold=100
```

```python
#brands which has less than 100 cars:
names=counts[counts<=threshold].index
names
```

```python
#performing encoding:
pd.get_dummies(df['brand'].replace(names,'uncommon')).astype(int)
```
