OneHotEncoding:

```python
#Importing Libraries:
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#importing dataframe:
df=pd.read_csv('cars.csv')
df.head(2)
```

OneHotEncoding using Pandas:

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

OneHotEncoding using Sklearn:

```python
#Train_test_split:

x=df.drop(columns=['selling_price'])
y=df['selling_price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.head(1)
```

```python
#defining column transformer:
col_tra=ColumnTransformer(transformers=[
    ('fuel',OneHotEncoder(drop='first',dtype=np.int32),['fuel']),
    ('owner',OneHotEncoder(drop='first',dtype=np.int32),['owner'])
],remainder='passthrough')
col_tra
```

```python
#Fitting data
x_train=col_tra.fit_transform(x_train)
x_test=col_tra.transform(x_test)
```

```python
x_train=pd.DataFrame(x_train, columns=col_tra.get_feature_names_out())
x_train.head()
```
